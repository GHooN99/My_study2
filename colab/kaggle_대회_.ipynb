```
import pandas as pd
import numpy as np
import seaborn as sns
```

```
data = pd.read_csv("train.csv")
data
```

```
data.info()
```

```
y_data = data.iloc[:,1]
y_data
```

```
data = data.drop(columns=['y','id','site','dadage'])
data
```

## missing value

dadage , momedu, site

```
data['mom.edu']=data['mom.edu'].fillna(method='backfill')
data.info()
```

```
#target=['bw',	'birth.o',	'momage',	'sex',	'twin',	'b.marr',	'mom.edu',	'cig'	,'first'	,'booze',	'drugs'	,'work.dur'	,'prenatal']
target=['bw','momage']
#normal
for k in target:
  data[k]=(data[k]-data[k].min())/(data[k].max()-data[k].min())
```

```
target=['b.head','preterm','nnhealth']
#std
for k in target:
  data[k]=(data[k]-data[k].mean())/(data[k].std())

```

```
model = LGBMRegressor(learning_rate=0.005,n_estimators=300,num_leaves=41)
model.fit(data,y_data)
```

## using DNN

```
from tensorflow import keras
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras import Sequential
from tensorflow.keras import optimizers
from tensorflow.keras.callbacks import EarlyStopping
```

```
modelDNN = Sequential()
modelDNN.add(Dense(32,input_dim = 15,activation='relu',kernel_initializer='he_normal',kernel_regularizer=keras.regularizers.l2(0.01)))
modelDNN.add(Dense(16,activation='relu',kernel_initializer='he_normal',kernel_regularizer=keras.regularizers.l2(0.01)))
modelDNN.add(Dense(8,activation='relu',kernel_initializer='he_normal',kernel_regularizer=keras.regularizers.l2(0.01)))
#modelDNN.add(Dense(4,activation='relu',kernel_initializer='he_normal',kernel_regularizer=keras.regularizers.l2(0.01)))
modelDNN.add(Dropout(0.2))
modelDNN.add(Dense(1,kernel_initializer='he_normal',kernel_regularizer=keras.regularizers.l2(0.01)))


early = EarlyStopping(patience=200)
optim = optimizers.Nadam(learning_rate=0.0026)
modelDNN.compile(optimizer=optim,loss='mse',metrics=['mse'])
modelDNN.fit(data,y_data,batch_size=16,epochs=1000,validation_split=0.15,callbacks=early)
```

```
k = modelDNN.predict(test_data)
k
```

```
resultDNN = pd.DataFrame({'id':test_id,'y':k.flatten()})
resultDNN
```
