```
import pandas as pd
import tensorflow as tf
from matplotlib import pyplot as plt
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

```
data = pd.read_csv("train.csv")
data.head(10)
```

```
x_data = data.iloc[:,1:-1]
x_data
```

```
y_data = data.iloc[:,-1]
y_data
```

```
with tf.device('/device:GPU:0'):
  model = Sequential()
  model.add(Dense(32,input_dim=8,activation='relu'))
  model.add(Dense(12,activation='relu'))
  model.add(Dense(6,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  optim = optimizers.Adam(learning_rate=0.0035)

  model.compile(optimizer=optim,loss='binary_crossentropy',metrics=['accuracy'])
  model.fit(x_data,y_data,batch_size=16,epochs=200)
```

```
test_data = pd.read_csv("test_data.csv")
test_data.head()
```

```
test_data = test_data.iloc[:,1:]
test_data.tail(5)
```

```
y_test = model.predict(test_data)
y_test.flatten()
y_test = np.where(y_test > 0.5, 1, 0)
```

```
result = pd.DataFrame({'ID':range(0,50),'Label':y_test.flatten()})
result
```

```
result.to_csv("result.csv",index=False,header=True)
```
